## Loading data + EDA

#### 1) "Merging" data from Demo and IoA datasets

```python
df = pd.read_excel('/content/gdrive/MyDrive/AI Notebooks/LPP/Data/IA_report.xlsx')
df.head(15)
```

```python
demo_df = pd.read_excel("/content/gdrive/MyDrive/AI Notebooks/LPP/Data/demo.xlsx")
demo_df.set_index('SubjectID', inplace=True)
demo_df.head()
```

```python
df[['Age', 'Sex(0-f,1-m)', 'IQ', 'English_Level']] = 0, 0, 0, 0
df.head(5)
```

```python
# Function to fill cell in a row with corresponding value from demo_df (SubjectID as a link).
# Takes row + name of column to fill, returns transformed row

def col_filler(x, col):
    res_series = x 
    res_series[col] = demo_df.loc[x.SubjectID, col]
    return res_series
```

```python
# Fill all columns we need with data from demographic dataset

df = df.apply(col_filler, axis=1, raw=False, col='Age')
df = df.apply(col_filler, axis=1, raw=False, col='Sex(0-f,1-m)')
df = df.apply(col_filler, axis=1, raw=False, col='IQ')
df = df.apply(col_filler, axis=1, raw=False, col='English_Level')
df.head(3)
```

```python
# Change order of columns to make dataframe look more convenient

cols = list(df.columns)
cols = [cols[0]] + [cols[-1]] + cols[-4:-1] + cols[1:-4]
df = df[cols]
df
```

```python
df.info()
```

```python
df[df['Sex(0-f,1-m)'].isna()]
```

```python
df = df.dropna(axis=0)
df.info()
```

```python
# Check columns for "." values

for col in df.columns:
    print(df[df[col] == '.'].shape[0], "\tdots in column", end='\t')
    print(col)
```

```python
# Look at values unable to cast into float in each column

def is_number(s):
    try:
        float(s)
        return True
    except:
        return False

for col in df.columns:
    print(col, ':', end=' ')
    for value in df[col].unique():
        if not(is_number(value)):
            print(value, end=' ')
    print()
```

Thus, the problematic cases are dots in several columns + dates in FIRST_SACCADE_AMPLITUDE

```python
# Drop every row containing "."

df.drop(df.loc[list(df.apply(lambda x: x == '.',  axis=1, raw=False).any(axis=1))].index)
```

```python
# Replace each non-numeric value with None

def check_col(col):
    if col.name not in ['SubjectID', 'English_Level']:
        return pd.to_numeric(col, errors='coerce')
    return col

df = df.apply(check_col, axis=0)
df.info()
```

## Machine Learning

#### Splitting data

```python
from sklearn import model_selection, preprocessing, pipeline, metrics
from sklearn import linear_model, neighbors, ensemble, naive_bayes, svm
```

```python
df.columns, len(df.columns)
```

```python
df_labels = df['English_Level']
df_data = df.drop(['English_Level'], axis=1)
print(f"df_data shape: {df_data.shape}")
print(f"df_labels shape: {df_labels.shape}")
```

```python
train_data, test_data, train_labels, test_labels = model_selection.train_test_split(df_data, df_labels)
print(f"train_data shape: {train_data.shape}")
print(f"train_labels shape: {train_labels.shape}")
print(f"test_data shape: {test_data.shape}")
print(f"test_labels shape: {test_labels.shape}")
```

```python
train_data.head()
```

```python
train_labels
```

```python
for col in train_data.columns:
    if(len(train_data[col].unique()) == 2):
        print(train_data.columns.get_loc(col))
```

```python
binary_cols = (2, 7, 14, 15, 16)
numeric_cols = (i for i in range(0, 18) if i not in binary_cols + [0,])
binary_cols_indexes = np.array([(train_data.columns.get_loc(col) in binary_cols) for col in train_data.columns], dtype = bool)
numeric_cols_indexes = np.invert(binary_cols_indexes)
numeric_cols_indexes[0] = np.False_

binary_cols_indexes, numeric_cols_indexes
```

### Using different classifiers

```python
'''
classifiers to run:

1) linear_model.LogisticRegressionCV(multi_class='multinomial', solver='saga')
2) ensemble.GradientBoostingClassifier()
3) ensemble.RandomForestClassifier()
4) neighbors.KNeighborsClassifier()
5) naive_bayes.GaussianNB()
6) svm.SVC()

'''
```

```python
def make_pipeline_estimator(classifier):

    pipe_estimator = pipeline.Pipeline(steps = [       
        ('feature_processing', pipeline.FeatureUnion(transformer_list = [        
                #binary features preprocessing (default)
                ('binary_variables_processing', preprocessing.FunctionTransformer(lambda data: data.loc[:, pd.Index(binary_cols_indexes)])), 
                    
                #numeric features preprocessing (scaling)
                ('numeric_variables_processing', pipeline.Pipeline(steps = [
                    ('selecting', preprocessing.FunctionTransformer(lambda data: data.loc[:, pd.Index(numeric_cols_indexes)])),
                    ('scaling', preprocessing.StandardScaler(with_mean = 0))            
                        ])),
            ])),
        ('model_fitting', classifier)
        ]
    )

    return pipe_estimator
```

```python
def draw_conf_matrix(test_labels, preds, header):
    plt.figure(figsize = (10,7))

    cm = metrics.confusion_matrix(test_labels, preds, labels=['C1', 'C2', 'B1', 'B2'])
    cm_df = pd.DataFrame(cm, index = ['C1', 'C2', 'B1', 'B2'], columns=['C1', 'C2', 'B1', 'B2'])
    sns.heatmap(cm_df, annot=True, fmt='d', annot_kws={"size" : 14}, cmap='Greens')
    plt.title(header, fontsize=14)
    plt.show()
```

#### 1) LogisticRegressionCV

```python
estimator = make_pipeline_estimator(linear_model.LogisticRegressionCV(multi_class='multinomial', solver='saga'))
estimator.fit(train_data, train_labels)
preds = estimator.predict(test_data)
```

```python
metrics.f1_score(test_labels, preds, average='weighted')
```

```python
draw_conf_matrix(test_labels, preds, "LogReg confusion matrix")
```

#### 2) Gradient Boosting

```python
estimator = make_pipeline_estimator(ensemble.GradientBoostingClassifier())
estimator.fit(train_data, train_labels)
preds = estimator.predict(test_data)
```

```python
metrics.f1_score(test_labels, preds, average='weighted')
```

```python
draw_conf_matrix(test_labels, preds, "GB confusion matrix")
```

#### 3) Random Forest

```python
estimator = make_pipeline_estimator(ensemble.RandomForestClassifier())
estimator.fit(train_data, train_labels)
preds = estimator.predict(test_data)
```

```python
metrics.f1_score(test_labels, preds, average='weighted')
```

```python
draw_conf_matrix(test_labels, preds, "RF confusion matrix")
```

#### 4) K Neighbors

```python
estimator = make_pipeline_estimator(neighbors.KNeighborsClassifier())
estimator.fit(train_data, train_labels)
preds = estimator.predict(test_data)
```

```python
metrics.f1_score(test_labels, preds, average='weighted')
```

```python
draw_conf_matrix(test_labels, preds, "KNeighdors confusion matrix")
```

#### 5) Gaussian Naive Bayes

```python
estimator = make_pipeline_estimator(naive_bayes.GaussianNB())
estimator.fit(train_data, train_labels)
preds = estimator.predict(test_data)
```

```python
metrics.f1_score(test_labels, preds, average='weighted')
```

```python
draw_conf_matrix(test_labels, preds, "Gaussian NB confusion matrix")
```

#### 6) Support Vectors Classifier

```python
estimator = make_pipeline_estimator(svm.SVC())
estimator.fit(train_data, train_labels)
preds = estimator.predict(test_data)
```

```python
metrics.f1_score(test_labels, preds, average='weighted')
```

```python
draw_conf_matrix(test_labels, preds, "SVC")
```
